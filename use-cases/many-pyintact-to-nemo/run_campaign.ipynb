````python
from pathlib import Path
from campaign_checks import load_json, run_all_checks, format_report

base = Path('example-output')
base.exists()
````

## Final quality gate (after tuning + rerun)

````python
manifest_final = load_json(base / 'campaign_jobs_final.json')
dataset_final = load_json(base / 'dataset_summary_final.json')
metrics_final = load_json(base / 'surrogate_metrics_final.json')

results_final = run_all_checks(manifest_final, dataset_final, metrics_final)
print(format_report(results_final))
````
